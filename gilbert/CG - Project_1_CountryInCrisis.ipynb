```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# import source data files

regions = pd.read_csv("raw_data/regions.csv")

od_2014 = pd.read_csv("raw_data/OD_2013_2014.csv")
od_2015 = pd.read_csv("raw_data/OD_2015.csv")
od_2016 = pd.read_csv("raw_data/OD_2016.csv")
od_2017 = pd.read_csv("raw_data/OD_2017.csv")

# import perscription data
rx_2013 = pd.read_csv("raw_data/RX_2013.csv")
rx_2014 = pd.read_csv("raw_data/RX_2014.csv")
rx_2015 = pd.read_csv("raw_data/RX_2015.csv")
rx_2016 = pd.read_csv("raw_data/RX_2016.csv")
rx_2017 = pd.read_csv("raw_data/RX_2017.csv")

# import population data
pop = pd.read_csv("raw_data/pop_2010-2018.csv")
```

```python
# rename columns for index on merge
od_2014 = od_2014.rename(columns={"2014Number":"OD_2014", "2013Number":"OD_2013"})
od_2015 = od_2015.rename(columns={"Number":"OD_2015"})
od_2016 = od_2016.rename(columns={"number":"OD_2016"})
od_2017 = od_2017.rename(columns={"number":"OD_2017"})
rx_2013 = rx_2013.rename(columns={"State":"State_Name", "State ABBR":"State", "2013 Prescribing Rate":"RX_2013"})
rx_2014 = rx_2014.rename(columns={"State":"State_Name", "State ABBR":"State", "2014 Prescribing Rate":"RX_2014"})
rx_2015 = rx_2015.rename(columns={"State":"State_Name", "State ABBR":"State", "2015 Prescribing Rate":"RX_2015"})
rx_2016 = rx_2016.rename(columns={"State":"State_Name", "State ABBR":"State", "2016 Prescribing Rate":"RX_2016"})
rx_2017 = rx_2017.rename(columns={"State":"State_Name", "Abbreviation":"State", "Year 2017":"RX_2017"})

pop = pop.rename(columns={"table with row headers in column A and column headers in rows 3 through 4. (leading dots indicate sub-parts)":"State",
                         'Unnamed: 6':'Pop_2013', 'Unnamed: 7':'Pop_2014', 'Unnamed: 8':'Pop_2015', 'Unnamed: 9':'Pop_2016', 'Unnamed: 10':'Pop_2017'})
pop = pop.replace({'.Alabama': 'AL', '.Alaska': 'AK', '.Arizona': 'AZ','.Arkansas': 'AR','.California': 'CA',
                    '.Colorado': 'CO','.Connecticut': 'CT','.Delaware': 'DE','.District of Columbia': 'DC',
                    '.Florida': 'FL','.Georgia': 'GA','.Hawaii': 'HI','.Idaho': 'ID','.Illinois': 'IL',
                    '.Indiana': 'IN','.Iowa': 'IA','.Kansas': 'KS','.Kentucky': 'KY','.Louisiana':'LA',
                    '.Maine': 'ME','.Maryland':'MD','.Massachusetts':'MA','.Michigan':'MI', '.Minnesota':'MN',
                    '.Mississippi':'MS','.Missouri':'MO','.Montana':'MT','.Nebraska':'NE','.Nevada':'NV',
                    '.New Hampshire':'NH','.New Jersey':'NJ','.New Mexico':'NM','.New York':'NY',
                    '.North Carolina':'NC','.North Dakota':'ND','.Ohio':'OH','.Oklahoma':'OK',
                    '.Oregon':'OR','.Pennsylvania':'PA','.Rhode Island':'RI','.South Carolina':'SC','.South Dakota':'SD',
                    '.Tennessee':'TN','.Texas':'TX','.Utah':'UT','.Vermont':'VT','.Virginia':'VA','.Washington':'WA',
                    '.West Virginia':'WV','.Wisconsin':'WI','.Wyoming':'WY'})
```

```python
# merge data for combined dataframe

master_df = pd.merge(regions, od_2014, on="State")
master_df = pd.merge(master_df, od_2015, on="State")
master_df = pd.merge(master_df, od_2016, on="State")
master_df = pd.merge(master_df, od_2017, on="State")
master_df = pd.merge(master_df, rx_2013, on="State")
master_df = pd.merge(master_df, rx_2014, on="State")
master_df = pd.merge(master_df, rx_2015, on="State")
master_df = pd.merge(master_df, rx_2016, on="State")
master_df = pd.merge(master_df, rx_2017, on="State")
master_df = pd.merge(master_df, pop, on="State")
```

```python
# drop unwanted columns
clean_df = master_df.drop(['2014Rate', '2014Range', '2013Rate',
        '2013Range', 'Change', 'Significant', 'Range', 'Rate',
        'range_x', 'rate_x', 'range_y', 'rate_y','State_Name_x', 'State_Name_y',
       'State_Name_x', 'State_Name_y', 'State_Name',
       'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4',
       'Unnamed: 5','Unnamed: 11'], axis=1)
```

```python
# re-order columns
final_df = clean_df[['Region','State','RX_2013', 'OD_2013', 'Pop_2013', 'RX_2014', 'OD_2014','Pop_2014', 
        'RX_2015', 'OD_2015', 'Pop_2015','RX_2016', 'OD_2016','Pop_2016', 'RX_2017','OD_2017','Pop_2017']]
```
