```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
master_df = pd.read_csv("Master_DF2.csv")
master_df.head()
```

```python
reg2013_RX_rate = master_df.groupby(['Region']).mean()['2013_RX_rate']
reg2014_RX_rate = master_df.groupby(['Region']).mean()['2014_RX_rate']
reg2015_RX_rate = master_df.groupby(['Region']).mean()['2015_RX_rate']
reg2016_RX_rate = master_df.groupby(['Region']).mean()['2016_RX_rate']
reg2017_RX_rate = master_df.groupby(['Region']).mean()['2017_RX_rate']
reg2013Deaths = master_df.groupby(['Region']).mean()['2013Deaths']
reg2014Deaths = master_df.groupby(['Region']).mean()['2014Deaths']
reg2015Deaths = master_df.groupby(['Region']).mean()['2015Deaths']
reg2016Deaths = master_df.groupby(['Region']).mean()['2016Deaths']
reg2017Deaths = master_df.groupby(['Region']).mean()['2017Deaths']
```

```python
reg_RX = pd.DataFrame({'2013_RX': reg2013_RX_rate, '2014_RX':reg2014_RX_rate, '2015_RX': reg2015_RX_rate,
                       '2016_RX': reg2016_RX_rate, '2017_RX': reg2017_RX_rate,})

reg_RX.head(10)
```

```python
reg_RX.loc['Far West', "2013_RX":"2017_RX"].plot(label="Far West")
reg_RX.loc['Great Lakes', "2013_RX":"2017_RX"].plot(label="Great Lakes")
reg_RX.loc['Mideast', "2013_RX":"2017_RX"].plot(label="Mideast")
reg_RX.loc['New England', "2013_RX":"2017_RX"].plot(label="New England")
reg_RX.loc['Plains', "2013_RX":"2017_RX"].plot(label="Plains")
reg_RX.loc['Rocky Mountain', "2013_RX":"2017_RX"].plot(label="Rocky Mountain")
reg_RX.loc['Southeast', "2013_RX":"2017_RX"].plot(label="Southeast")
reg_RX.loc['Southwest', "2013_RX":"2017_RX"].plot(label="Southwest")


plt.title("Average Opioid RX by Region")
plt.xlabel("Year")
plt.ylabel("Prescriptions per 100 people")
plt.grid()

plt.legend()
plt.savefig("regRX.png")
plt.show()
```

```python
reg_death = pd.DataFrame({'2013_Deaths': reg2013Deaths, '2014_Deaths': reg2014Deaths, '2015_Deaths': reg2015Deaths,
                      '2016_Deaths': reg2016Deaths,'2017_Deaths': reg2017Deaths})

reg_death.head(10)
```

```python
reg_death.loc['Far West', "2013_Deaths":"2017_Deaths"].plot(label="Far West")
reg_death.loc['Great Lakes', "2013_Deaths":"2017_Deaths"].plot(label="Great Lakes")
reg_death.loc['Mideast', "2013_Deaths":"2017_Deaths"].plot(label="Mideast")
reg_death.loc['New England', "2013_Deaths":"2017_Deaths"].plot(label="New England")
reg_death.loc['Plains', "2013_Deaths":"2017_Deaths"].plot(label="Plains")
reg_death.loc['Rocky Mountain', "2013_Deaths":"2017_Deaths"].plot(label="Rocky Mountain")
reg_death.loc['Southeast', "2013_Deaths":"2017_Deaths"].plot(label="Southeast")
reg_death.loc['Southwest', "2013_Deaths":"2017_Deaths"].plot(label="Southwest")


plt.title("Average Opioid Deaths by Region")
plt.xlabel("Year")
plt.ylabel("Deaths per 100,000 people")
plt.grid()
plt.legend()
plt.savefig("regdeath.png")
plt.show()
```
