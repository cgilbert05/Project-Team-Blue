```python
# import dependencies
%matplotlib notebook
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# import source data files
rx_v_deaths = pd.read_csv("Master_DF2.csv")
rx_v_deaths.head()
```

```python
# Average of deaths each year
mean_deaths_2013 = rx_v_deaths["2013Deaths"].mean()
mean_deaths_2014 = rx_v_deaths["2014Deaths"].mean()
mean_deaths_2015 = rx_v_deaths["2015Deaths"].mean()
mean_deaths_2016 = rx_v_deaths["2016Deaths"].mean()
mean_deaths_2017 = rx_v_deaths["2017Deaths"].mean()

# Averge prescription rates each year
mean_rx_2013 = rx_v_deaths["2013_RX_rate"].mean()
mean_rx_2014 = rx_v_deaths["2014_RX_rate"].mean()
mean_rx_2015 = rx_v_deaths["2015_RX_rate"].mean()
mean_rx_2016 = rx_v_deaths["2016_RX_rate"].mean()
mean_rx_2017 = rx_v_deaths["2017_RX_rate"].mean()
```

```python
avg_death = [mean_deaths_2013, mean_deaths_2014, mean_deaths_2015, mean_deaths_2016, mean_deaths_2017]
avg_rx = [mean_rx_2013, mean_rx_2014, mean_rx_2015, mean_rx_2016, mean_rx_2017]
time = [2013, 2014, 2015, 2016, 2017]
avg_death_data = plt.plot(time, avg_death, color="red", label="Death rates")
avg_rx_data = plt.plot(time, avg_rx, color = "blue", label="Prescription rates")

plt.title("Average Death v Average Prescription Rates")
plt.xlabel("Year")
plt.ylabel("Rates")

plt.xlim(2013,2017)
plt.ylim(0,100)

plt.grid()
plt.show()
```

```python
# Sum of deaths 

#southeast
se_deaths_2013 = southeast["2013Deaths"].mean()
se_deaths_2014 = southeast["2014Deaths"].mean()
se_deaths_2015 = southeast["2015Deaths"].mean()
se_deaths_2016 = southeast["2016Deaths"].mean()
se_deaths_2017 = southeast["2017Deaths"].mean()

#southwest
sw_deaths_2013 = southwest["2013Deaths"].mean()
sw_deaths_2014 = southwest["2014Deaths"].mean()
sw_deaths_2015 = southwest["2015Deaths"].mean()
sw_deaths_2016 = southwest["2016Deaths"].mean()
sw_deaths_2017 = southwest["2017Deaths"].mean()

#farwest
fw_deaths_2013 = farwest["2013Deaths"].mean()
fw_deaths_2014 = farwest["2014Deaths"].mean()
fw_deaths_2015 = farwest["2015Deaths"].mean()
fw_deaths_2016 = farwest["2016Deaths"].mean()
fw_deaths_2017 = farwest["2017Deaths"].mean()

#rockymountain
rm_deaths_2013 = rockmount["2013Deaths"].mean()
rm_deaths_2014 = rockmount["2014Deaths"].mean()
rm_deaths_2015 = rockmount["2015Deaths"].mean()
rm_deaths_2016 = rockmount["2016Deaths"].mean()
rm_deaths_2017 = rockmount["2017Deaths"].mean()

#plains
p_deaths_2013 = plains["2013Deaths"].mean()
p_deaths_2014 = plains["2014Deaths"].mean()
p_deaths_2015 = plains["2015Deaths"].mean()
p_deaths_2016 = plains["2016Deaths"].mean()
p_deaths_2017 = plains["2017Deaths"].mean()

#greatlakes
gl_deaths_2013 = greatlakes["2013Deaths"].mean()
gl_deaths_2014 = greatlakes["2014Deaths"].mean()
gl_deaths_2015 = greatlakes["2015Deaths"].mean()
gl_deaths_2016 = greatlakes["2016Deaths"].mean()
gl_deaths_2017 = greatlakes["2017Deaths"].mean()

#mideast
me_deaths_2013 = mideast["2013Deaths"].mean()
me_deaths_2014 = mideast["2014Deaths"].mean()
me_deaths_2015 = mideast["2015Deaths"].mean()
me_deaths_2016 = mideast["2016Deaths"].mean()
me_deaths_2017 = mideast["2017Deaths"].mean()

#newengland
ne_deaths_2013 = newengland["2013Deaths"].mean()
ne_deaths_2014 = newengland["2014Deaths"].mean()
ne_deaths_2015 = newengland["2015Deaths"].mean()
ne_deaths_2016 = newengland["2016Deaths"].mean()
ne_deaths_2017 = newengland["2017Deaths"].mean()

# Sum of Prescriptions

#southeast
se_rx_2013 = southeast["2013_RX_rate"].mean()
se_rx_2014 = southeast["2014_RX_rate"].mean()
se_rx_2015 = southeast["2015_RX_rate"].mean()
se_rx_2016 = southeast["2016_RX_rate"].mean()
se_rx_2017 = southeast["2017_RX_rate"].mean()


#southwest
sw_rx_2013 = southwest["2013_RX_rate"].mean()
sw_rx_2014 = southwest["2014_RX_rate"].mean()
sw_rx_2015 = southwest["2015_RX_rate"].mean()
sw_rx_2016 = southwest["2016_RX_rate"].mean()
sw_rx_2017 = southwest["2017_RX_rate"].mean()

#farwest
fw_rx_2013 = farwest["2013_RX_rate"].mean()
fw_rx_2014 = farwest["2014_RX_rate"].mean()
fw_rx_2015 = farwest["2015_RX_rate"].mean()
fw_rx_2016 = farwest["2016_RX_rate"].mean()
fw_rx_2017 = farwest["2017_RX_rate"].mean()

#rockymountain
rm_rx_2013 = rockmount["2013_RX_rate"].mean()
rm_rx_2014 = rockmount["2014_RX_rate"].mean()
rm_rx_2015 = rockmount["2015_RX_rate"].mean()
rm_rx_2016 = rockmount["2016_RX_rate"].mean()
rm_rx_2017 = rockmount["2017_RX_rate"].mean()

#plains
p_rx_2013 = plains["2013_RX_rate"].mean()
p_rx_2014 = plains["2014_RX_rate"].mean()
p_rx_2015 = plains["2015_RX_rate"].mean()
p_rx_2016 = plains["2016_RX_rate"].mean()
p_rx_2017 = plains["2017_RX_rate"].mean()

#greatlakes
gl_rx_2013 = greatlakes["2013_RX_rate"].mean()
gl_rx_2014 = greatlakes["2014_RX_rate"].mean()
gl_rx_2015 = greatlakes["2015_RX_rate"].mean()
gl_rx_2016 = greatlakes["2016_RX_rate"].mean()
gl_rx_2017 = greatlakes["2017_RX_rate"].mean()

#mideast
me_rx_2013 = mideast["2013_RX_rate"].mean()
me_rx_2014 = mideast["2014_RX_rate"].mean()
me_rx_2015 = mideast["2015_RX_rate"].mean()
me_rx_2016 = mideast["2016_RX_rate"].mean()
me_rx_2017 = mideast["2017_RX_rate"].mean()

#newengland
ne_rx_2013 = newengland["2013_RX_rate"].mean()
ne_rx_2014 = newengland["2014_RX_rate"].mean()
ne_rx_2015 = newengland["2015_RX_rate"].mean()
ne_rx_2016 = newengland["2016_RX_rate"].mean()
ne_rx_2017 = newengland["2017_RX_rate"].mean()
```

```python
# Total sum of deaths over 5 year period by region
se_death_sum = (se_deaths_2013 + se_deaths_2014 + se_deaths_2015 + se_deaths_2016 + se_deaths_2017)/5
sw_death_sum = (sw_deaths_2013 + sw_deaths_2014 + sw_deaths_2015 + sw_deaths_2016 + sw_deaths_2017)/5
fw_death_sum = (fw_deaths_2013 + fw_deaths_2014 + fw_deaths_2015 + fw_deaths_2016 + fw_deaths_2017)/5
rm_death_sum = (rm_deaths_2013 + rm_deaths_2014 + rm_deaths_2015 + rm_deaths_2016 + rm_deaths_2017)/5
p_death_sum = (p_deaths_2013 + p_deaths_2014 + p_deaths_2015 + p_deaths_2016 + p_deaths_2017)/5
gl_death_sum = (gl_deaths_2013 + gl_deaths_2014 + gl_deaths_2015 + gl_deaths_2016 + gl_deaths_2017)/5
me_death_sum = (me_deaths_2013 + me_deaths_2014 + me_deaths_2015 + me_deaths_2016 + me_deaths_2017)/5
ne_death_sum = (ne_deaths_2013 + ne_deaths_2014 + ne_deaths_2015 + ne_deaths_2016 + ne_deaths_2017)/5

# Total sum of prescription rates over 5 year period by region
se_rx_sum = (se_rx_2013 + se_rx_2014 + se_rx_2015 + se_rx_2016 + se_rx_2017)/5
sw_rx_sum = (sw_rx_2013 + sw_rx_2014 + sw_rx_2015 + sw_rx_2016 + sw_rx_2017)/5
fw_rx_sum = (fw_rx_2013 + fw_rx_2014 + fw_rx_2015 + fw_rx_2016 + fw_rx_2017)/5
rm_rx_sum = (rm_rx_2013 + rm_rx_2014 + rm_rx_2015 + rm_rx_2016 + rm_rx_2017)/5
p_rx_sum = (p_rx_2013 + p_rx_2014 + p_rx_2015 + p_rx_2016 + p_rx_2017)/5
gl_rx_sum = (gl_rx_2013 + gl_rx_2014 + gl_rx_2015 + gl_rx_2016 + gl_rx_2017)/5
me_rx_sum = (me_rx_2013 + me_rx_2014 + me_rx_2015 + me_rx_2016 + me_rx_2017)/5
ne_rx_sum = (ne_rx_2013 + ne_rx_2014 + ne_rx_2015 + ne_rx_2016 + ne_rx_2017)/5
```

```python
death_rates = [se_death_sum, sw_death_sum, fw_death_sum, rm_death_sum, p_death_sum, gl_death_sum, me_death_sum, ne_death_sum]
rx_rates = [se_rx_sum, sw_rx_sum, fw_rx_sum, rm_rx_sum, p_rx_sum, gl_rx_sum, me_rx_sum, ne_rx_sum]
x_axis = np.arange(len(death_rates))

plt.bar(x_axis+0, death_rates, color='red', width=0.4)
plt.bar(x_axis+0.4, rx_rates, color="blue", width=0.4)

plt.title("Average Death and Rx Rates by Region")
plt.xlabel("Region")
plt.ylabel("Rates")

tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Southeast", "Southwest", "Far West", "Rocky Mountain", "Plains", "Great Lakes", "Mideast", "New England"], rotation = 25)

plt.xlim(-0.5, len(x_axis)-0.25)
plt.ylim(0,max(rx_rates)+20)
Rates = ["Death", "RX"]
plt.legend(Rates)

plt.savefig("DeathandRxTotals.png")
plt.show()
```
